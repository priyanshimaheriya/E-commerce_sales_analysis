```python
# Filter the dataset for "Women" and "Men"
women_retail = df[df['Gender'] == 'Women']
men_retail = df[df['Gender'] == 'Men']

# Count occurrences of each primary color for "Women"
women_color_counts = women_retail['PrimaryColor'].value_counts().head(10)

# Count occurrences of each primary color for "Men"
men_color_counts = men_retail['PrimaryColor'].value_counts().head(10)

# Combine the color counts into a single DataFrame
color_counts_df = pd.DataFrame({
    'Women': women_color_counts,
    'Men': men_color_counts
}).fillna(0)  # Fill NaN values with 0 for colors not present in both genders

print(color_counts_df)
print("----------------------------")

# Plot the grouped bar graph
color_counts_df.plot(kind='bar', figsize=(14, 8), color=['#B73377', '#6DD2FF'], width=0.8)
plt.title('Primary Color Counts for Women and Men')
plt.xlabel('Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```

For defining the primary colors counts for women and men the for women least and highest color selected is Brown and Blue and on the oppposite for men is Maroon and Blue. At a certain point Black color stands as nearly equal choice of selection for both the gender.

## What is the price distribution of products sold on Myntra, and which price ranges are the most popular?

```python
# Define price ranges for grouping
price_bins = [0, 500, 1000, 1500, 2000, 3000, 5000, 10000, 20000, 30000, 50000, float('inf')]
price_labels = ['<500', '500-1000', '1000-1500', '1500-2000', '2000-3000', '3000-5000', '5000-10000', 
                '10000-20000', '20000-30000', '30000-50000', '>50000']

# Bin the 'Price (INR)' data
df['PriceRange'] = pd.cut(df['Price (INR)'], bins=price_bins, labels=price_labels, right=False)

# Count the number of products in each price range
price_range_counts = df['PriceRange'].value_counts().sort_index()


print(price_range_counts)
print('---------------------------')


# Plotting the data
plt.figure(figsize=(12, 6))
price_range_counts.plot(kind='line', marker='o')

plt.title('Number of Products in Each Price Range')
plt.xlabel('Price Range')
plt.ylabel('Number of Products')
plt.xticks(range(len(price_labels)), price_labels, rotation=45)
plt.tight_layout()
plt.grid()

plt.show()
```

The price distribution in myntra range starts from 500 to 50000 from which the most popular price range where people buy the most is between 500-1000. Also the products available for that range is upto 5659 which is the highest amongst all. But for the range 50000 and above there is bare minimum number of products available which is only 4 items in the list. On the Average number of products between 1000 to 3000 there is average number of price range from 1000 to 3000.

## Brand popularity analysis : 

```python

top_brands = df['ProductBrand'].value_counts().head(30).index
brand_counts = df['ProductBrand'].value_counts().head(30)

# Calculate average price for each of the top 20 brands
avg_prices = df[df['ProductBrand'].isin(top_brands)].groupby('ProductBrand')['Price (INR)'].mean().reindex(top_brands)


count_and_avg = pd.concat([brand_counts,avg_prices],axis=1)
print(count_and_avg.rename(columns={'count':'Brand_Counts','Price (INR)':'Avg_Price'}))

# Plotting the data
fig, ax1 = plt.subplots(figsize=(12, 8))

# Bar graph for number of products
color="lightgreen"
ax1.set_xlabel('Brand')
ax1.set_ylabel('Number of Products', color='#19aade')
ax1.bar(brand_counts.index, brand_counts.values, color=color)
ax1.tick_params(axis='y', labelcolor='#19aade')
ax1.set_xticks(range(len(brand_counts.index))) 
ax1.set_xticklabels(brand_counts.index, rotation=45, ha='right')


# Line graph for average price
ax2 = ax1.twinx()
color = '#142459'
ax2.set_ylabel('Average Price (INR)', color='#213B91')
ax2.plot(avg_prices.index, avg_prices.values, color=color, marker='o',markerfacecolor='r')
ax2.tick_params(axis='y', labelcolor='#213B91')

# Title and layout
plt.title('Top 20 Brands by Popularity')
fig.tight_layout()

# Show the plot
plt.show()
```

In top 20 brands Indian Terrain brands sells more products with 971 Product counts at average price 966.057673 INR. And last in top 20 is SHISEIDO with 77 product count at average price 2203.376623 INR.
By the Average Price top Brand is Titan at 5918.242991 INR with 107 product count.
